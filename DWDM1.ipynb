```python
# Import the apyori library
from apyori import apriori

# Define the dataset
dataset = [
    ['apple', 'banana', 'orange'],
    ['banana', 'grape', 'watermelon'],
    ['apple', 'banana', 'grape'],
    ['banana', 'grape'],
    ['apple', 'banana', 'grape', 'watermelon']
]

# Define the minimum support, confidence and lift values
min_support = 0.4
min_confidence = 0.8
min_lift = 1.0

# Apply the Apriori algorithm
rules = apriori(dataset, min_support=min_support, min_confidence=min_confidence, min_lift=min_lift)

# Print the results
for rule in rules:
    print(rule)
    print("\n")
```

```python
# Import the pyfpgrowth library
import pyfpgrowth

# Define the dataset
dataset = [
    ['apple', 'banana', 'orange'],
    ['banana', 'grape', 'watermelon'],
    ['apple', 'banana', 'grape'],
    ['banana', 'grape'],
    ['apple', 'banana', 'grape', 'watermelon']
]

# Define the minimum support value
min_support = 2

# Apply the FP-Growth algorithm
patterns = pyfpgrowth.find_frequent_patterns(dataset, min_support)

# Generate the association rules
rules = pyfpgrowth.generate_association_rules(patterns, 0.8)

# Print the results
print("Frequent itemsets: \n", patterns)
print("Association rules: \n", rules)
```

```python
# Import the necessary libraries
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 

# Load the dataset

iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7)

# Define the decision tree classifier
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)

# Train the classifier using the training data
clf.fit(X_train, y_train)

# Predict the test data using the trained classifier
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
# Import the necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score

# Load the dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Define the SVM classifier
clf = svm.SVC(kernel='linear', C=1)

# Train the classifier using the training data
clf.fit(X_train, y_train)

# Predict the test data using the trained classifier
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
# Import the necessary libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Define the KNN classifier
k = 5
clf = KNeighborsClassifier(n_neighbors=k)

# Train the classifier using the training data
clf.fit(X_train, y_train)

# Predict the test data using the trained classifier
y_pred = clf.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris

# Load the Iris dataset
data = load_iris()
X = data.data
y = data.target

# Split the dataset into training and tes ting sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train the Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# Predict the test set labels
y_pred = model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt


iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)


X = iris.iloc[:, :-1].values


# wcss = []
# for i in range(1, 11):
#     kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
#     kmeans.fit(X)
#     wcss.append(kmeans.inertia_)
# plt.plot(range(1, 11), wcss)
# plt.title('Elbow Method')
# plt.xlabel('Number of clusters')
# plt.ylabel('WCSS')
# plt.show()


kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
y_kmeans = kmeans.fit_predict(X)


plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'yellow', label = 'Centroids')
plt.title('Iris Clusters')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend()
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt


iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)


X = iris.iloc[:, :-1].values


cluster = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
y_pred = cluster.fit_predict(X)


plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.title('Iris Clusters')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.legend()
plt.show()
```

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.pyplot as plt


iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)


X = iris.iloc[:, :-1].values
distance = 'euclidean'
linkage_method = 'ward'


Z = linkage(X, linkage_method, metric=distance)


plt.figure(figsize=(25, 10))
dendrogram(Z)
plt.title('Iris Hierarchical Clustering Dendrogram')
plt.xlabel('Data points')
plt.ylabel('Distance')
plt.show()
```
